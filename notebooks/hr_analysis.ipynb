```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\ya465\Desktop\Data Analyst\end-to-end\data\WA_Fn-UseC_-HR-Employee-Attrition.csv")
print(f'df_Shape: {df.shape}')
df.head()
```

----

* Basic Overview

```python
print("Shape:", df.shape)
print("\nColumn Names:\n", df.columns.tolist())
print("\nData Types:\n", df.dtypes)
print("\nMissing Values:\n", df.isnull().sum())
print("\nAttrition Value Counts:\n", df['Attrition'].value_counts())
print("\nBasic Stats:")
df.describe()
```

-----

* Data Cleaning

```python
# removing unwanted calumns from df
df.drop(columns=['EmployeeCount', 'EmployeeNumber', 'Over18', 'StandardHours'], inplace= True)
df.head(2)
```

```python
#Attrittion Standarzition
df['Attrition'] = df['Attrition'].map({'Yes' : 1, 'No' : 0})
```

```python
df.head(2)
```

```python
# OveverTime 
df['OverTime'] = df['OverTime'].map({'Yes': 1, 'No': 0})
df[['EducationField','OverTime']].head(5)
```

---

* Adding Conditional Columns

```python
# age groups
df['Age'].unique()
```

```python
df['AgeGroup'] = pd.cut(df['Age'], bins= [18,25,35,45,60], labels= ['18-25', '26-35', '36-45', '45+'])
df[['Age','AgeGroup']].head(3) # Assigning Age Groups to the Employees
```

```python
# SalaryBrackets
df['SalaryBand'] = pd.cut(df['MonthlyIncome'], bins= [0, 3000, 7000, 20000], labels= ['Low', 'Mid' , 'High'])
df[['MonthlyIncome', 'SalaryBand']].tail(5)
```

```python
# TenureBracket
df['TenureBracket'] = pd.cut(df['YearsAtCompany'], bins= [-1,2,5,10,40], labels= ['0-2 yrs', '3-5 yrs', '6-10 yrs', '10+ yrs'])
df[['YearsAtCompany', 'TenureBracket']].head(4)
```

```python
# Export cleaned CSV
df.to_csv(r'C:\Users\ya465\Desktop\Data Analyst\end-to-end\data\hr_cleaned.csv', index=False)
print(f'hr_cleaned.csv Saved')
```

# EDA - Visualizations

```python
# importing visualizations libraries
import matplotlib.pyplot as plt
import seaborn as sns

# styling

sns.set_theme(style = 'whitegrid')
plt.rcParams['figure.figsize'] = (10,5)
plt.rcParams['font.family'] = 'sans-serif'

print('all action done')
```

```python
# Chart 1 - Attrition Distribution
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Count plot
attrition_counts = df['Attrition'].value_counts()
axes[0].bar(['Retained (0)', 'Attrited (1)'], attrition_counts.values, 
            color=['#2E75B6', '#E74C3C'], edgecolor='white', linewidth=1.5)
axes[0].set_title('Employee Attrition Count', fontsize=14, fontweight='bold')
axes[0].set_ylabel('Number of Employees')
for i, v in enumerate(attrition_counts.values):
    axes[0].text(i, v + 10, str(v), ha='center', fontweight='bold', fontsize=12)

# Pie chart
axes[1].pie(attrition_counts.values, labels=['Retained', 'Attrited'], 
            colors=['#2E75B6', '#E74C3C'], autopct='%1.1f%%', startangle=90,
            textprops={'fontsize': 12})
axes[1].set_title('Attrition Rate %', fontsize=14, fontweight='bold')

plt.suptitle('Overall Attrition Overview', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig(r'C:\Users\ya465\Desktop\Data Analyst\end-to-end\charts\chart1_attrition_overview.png', 
            dpi=150, bbox_inches='tight')
plt.show()
print("Chart 1 saved!")
```

```python
# Chart 2 - Attrition by Department
fig, ax = plt.subplots(figsize=(10, 5))

dept_attrition = df.groupby('Department')['Attrition'].mean() * 100

bars = ax.bar(dept_attrition.index, dept_attrition.values,
              color=['#E74C3C', '#2E75B6', '#F39C12'], 
              edgecolor='white', linewidth=1.5, width=0.5)

ax.set_title('Attrition Rate by Department (%)', fontsize=14, fontweight='bold')
ax.set_ylabel('Attrition Rate (%)')
ax.set_xlabel('Department')

for bar, val in zip(bars, dept_attrition.values):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.3,
            f'{val:.1f}%', ha='center', fontweight='bold', fontsize=12)

plt.tight_layout()
plt.savefig(r'C:\Users\ya465\Desktop\Data Analyst\end-to-end\charts\chart2_attrition_by_dept.png',
            dpi=150, bbox_inches='tight')
plt.show()
print("Chart 2 saved!")
```

```python
# Chart 3 - Attrition by Age Group
fig, ax = plt.subplots(figsize=(10, 5))

age_attrition = df.groupby('AgeGroup', observed=True)['Attrition'].mean() * 100

bars = ax.bar(age_attrition.index, age_attrition.values,
              color=['#E74C3C', '#E67E22', '#2E75B6', '#27AE60'],
              edgecolor='white', linewidth=1.5, width=0.5)

ax.set_title('Attrition Rate by Age Group (%)', fontsize=14, fontweight='bold')
ax.set_ylabel('Attrition Rate (%)')
ax.set_xlabel('Age Group')

for bar, val in zip(bars, age_attrition.values):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.3,
            f'{val:.1f}%', ha='center', fontweight='bold', fontsize=12)

plt.tight_layout()
plt.savefig(r'C:\Users\ya465\Desktop\Data Analyst\end-to-end\charts\chart3_attrition_by_age.png',
            dpi=150, bbox_inches='tight')
plt.show()
print("Chart 3 saved!")
```

```python
# Chart 4 - Monthly Income by Job Role
fig, ax = plt.subplots(figsize=(14, 6))

job_order = df.groupby('JobRole')['MonthlyIncome'].median().sort_values(ascending=False).index

sns.boxplot(data=df, x='JobRole', y='MonthlyIncome', order=job_order,
            palette='Blues_d', ax=ax)

ax.set_title('Monthly Income Distribution by Job Role', fontsize=14, fontweight='bold')
ax.set_ylabel('Monthly Income (USD)')
ax.set_xlabel('')
ax.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.savefig(r'C:\Users\ya465\Desktop\Data Analyst\end-to-end\charts\chart4_income_by_jobrole.png',
            dpi=150, bbox_inches='tight')
plt.show()
print("Chart 4 saved!")
```

```python
# Chart 5 - Overtime vs Attrition
fig, ax = plt.subplots(figsize=(8, 5))

overtime_attrition = df.groupby('OverTime')['Attrition'].mean() * 100

bars = ax.bar(['No Overtime (0)', 'Overtime (1)'], overtime_attrition.values,
              color=['#2E75B6', '#E74C3C'],
              edgecolor='white', linewidth=1.5, width=0.4)

ax.set_title('Attrition Rate — Overtime vs No Overtime (%)', fontsize=14, fontweight='bold')
ax.set_ylabel('Attrition Rate (%)')
ax.set_xlabel('Overtime Status')

for bar, val in zip(bars, overtime_attrition.values):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.3,
            f'{val:.1f}%', ha='center', fontweight='bold', fontsize=13)

plt.tight_layout()
plt.savefig(r'C:\Users\ya465\Desktop\Data Analyst\end-to-end\charts\chart5_overtime_attrition.png',
            dpi=150, bbox_inches='tight')
plt.show()
print("Chart 5 saved!")
```

```python
# Chart 6 - Correlation Heatmap
fig, ax = plt.subplots(figsize=(14, 10))

# Only numeric columns
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
corr_matrix = df[numeric_cols].corr()

sns.heatmap(corr_matrix, 
            annot=True, 
            fmt='.2f', 
            cmap='coolwarm',
            center=0,
            linewidths=0.5,
            annot_kws={'size': 8},
            ax=ax)

ax.set_title('Correlation Heatmap — All Numeric Features', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig(r'C:\Users\ya465\Desktop\Data Analyst\end-to-end\charts\chart6_correlation_heatmap.png',
            dpi=150, bbox_inches='tight')
plt.show()
print("Chart 6 saved!")
```

```python
# Chart 7 - Attrition by Salary Band
fig, ax = plt.subplots(figsize=(8, 5))

salary_attrition = df.groupby('SalaryBand', observed=True)['Attrition'].mean() * 100

bars = ax.bar(salary_attrition.index, salary_attrition.values,
              color=['#E74C3C', '#F39C12', '#27AE60'],
              edgecolor='white', linewidth=1.5, width=0.4)

ax.set_title('Attrition Rate by Salary Band (%)', fontsize=14, fontweight='bold')
ax.set_ylabel('Attrition Rate (%)')
ax.set_xlabel('Salary Band')

for bar, val in zip(bars, salary_attrition.values):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.3,
            f'{val:.1f}%', ha='center', fontweight='bold', fontsize=13)

plt.tight_layout()
plt.savefig(r'C:\Users\ya465\Desktop\Data Analyst\end-to-end\charts\chart7_attrition_by_salary.png',
            dpi=150, bbox_inches='tight')
plt.show()
print("Chart 7 saved!")
```
